Mini-projoect marks adding

```python
import csv

file=open('marks.csv','r')
csvreader=csv.reader(file)
header=next(csvreader) # extracts the header
print(header)

totalMarks={}
for row in csvreader:
  x=row[0]
  if x in totalMarks:
    totalMarks[x]=totalMarks[x]+int(row[1])
  else:
    totalMarks[x]=int(row[1])

print(totalMarks)
```

```python
import pandas as pd
df=pd.read_csv('/content/marks.csv')
df.head()
```

```python
df['Roll Num'].unique()
```

```python
roll_no=list(df['Roll Num'].unique())

output = {}

for r, m in zip(df['Roll Num'], df['Marks']):
    if r not in output:
        output[r] = 0
    output[r] += m

print(output)
```

```python
Output=pd.DataFrame(list(output.items()),columns=['Roll Num','Total Marks'])

Output.head()
```
